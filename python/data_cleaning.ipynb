```python
import pandas as pd

df = pd.read_csv("raw data.csv.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['created_date'] = pd.to_datetime(df['created_date'])
df['resolved_date'] = pd.to_datetime(df['resolved_date'], errors='coerce')
```

```python
df['resolution_time_hours'] = (
    (df['resolved_date'] - df['created_date']).dt.total_seconds() / 3600
)
df['resolution_time_hours'] = df['resolution_time_hours'].fillna(0)
```

```python
sla_mapping = {
    'Critical': 4,
    'High': 8,
    'Medium': 24,
    'Low': 48
}

df['sla_hours'] = df['priority'].map(sla_mapping)
```

```python
import numpy as np

df['sla_status'] = np.where(
    df['resolution_time_hours'] > df['sla_hours'],
    'Violated',
    'Met'
)
```

```python
df[['priority', 'resolution_time_hours', 'sla_hours', 'sla_status']].head()
```

```python
df.to_csv("cleaned_helpdesk_data.csv", index=False)
```
